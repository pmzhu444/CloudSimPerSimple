```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
df = pd.read_csv("../file/8_17/test.csv", sep=",")
df = df.drop(['Unnamed: 16'], axis=1)  
df
```

```python
df_1 = df.loc[df['dcName'] == 'DC_1'].reset_index()
df_2 = df.loc[df['dcName'] == 'DC_2'].reset_index()
df_3 = df.loc[df['dcName'] == 'DC_3'].reset_index()
```

```python
df_ReSupply = pd.DataFrame({'1':df_1['reSupply'], '2':df_2['reSupply'], '3':df_3['reSupply']})
```
