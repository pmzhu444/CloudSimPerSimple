```python
import numpy as np
import pandas as pd

import matplotlib
#matplotlib.rcParams['backend'] = 'SVG'
import matplotlib.pyplot as plt

plt.rc('font',family='Times New Roman')
```

```python
names = ['CR', 'OS', 'MB', 'SH', 'LA']
```

rr_data

```python
df_rr = pd.read_csv("./file/4_11/rr.csv", sep=",")
df_rr = df_rr.drop(['Unnamed: 15'], axis=1)

df_CR_rr = df_rr.loc[df_rr['dcName'] == 'DC_CR'].reset_index()
df_OS_rr = df_rr.loc[df_rr['dcName'] == 'DC_OS'].reset_index()
df_SH_rr = df_rr.loc[df_rr['dcName'] == 'DC_SH'].reset_index()
df_MB_rr = df_rr.loc[df_rr['dcName'] == 'DC_MB'].reset_index()
df_LA_rr = df_rr.loc[df_rr['dcName'] == 'DC_LA'].reset_index()

df_DcConsume_rr = pd.DataFrame({'CR':df_CR_rr['dcConsume'], 'MB':df_MB_rr['dcConsume'], 'OS':df_OS_rr['dcConsume'], 'SH':df_SH_rr['dcConsume'], 'LA':df_LA_rr['dcConsume']})
df_BrownConsume_rr = pd.DataFrame({'CR':df_CR_rr['dcPrice'], 'MB':df_MB_rr['dcPrice'], 'OS':df_OS_rr['dcPrice'], 'SH':df_SH_rr['dcPrice'], 'LA':df_LA_rr['dcPrice']})
df_BrownConsume_rr = df_BrownConsume_rr * 2
df_ReDcUsage_rr = pd.DataFrame({'CR':df_CR_rr['reDcUsage'], 'OS':df_OS_rr['reDcUsage'], 'MB':df_MB_rr['reDcUsage'], 'SH':df_SH_rr['reDcUsage'], 'LA':df_LA_rr['reDcUsage']})
df_ReConsume_rr = pd.DataFrame({'CR':df_CR_rr['reSupply'], 'MB':df_MB_rr['reSupply'], 'OS':df_OS_rr['reSupply'], 'SH':df_SH_rr['reSupply'], 'LA':df_LA_rr['reSupply']})
for name in names:
    for i in range(len(df_ReConsume_rr[name])):
        df_ReConsume_rr[name][i] = df_ReConsume_rr[name][i] * df_ReDcUsage_rr[name][i]
```
