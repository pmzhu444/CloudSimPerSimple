```python
import numpy as np
import pandas as pd

from matplotlib import pyplot as plt
```

```python
name_list = ['DC_MB', 'DC_SH', 'DC_OS', 'DC_LA', 'DC_CR']
idle = [28440, 24840, 22980, 21240, 19740]
full = [61800, 54600, 54000, 51300, 45900]
x =list(range(len(name_list)))
total_width, n = 0.8, 2
width = total_width / n
fig = plt.figure(figsize=(8, 5)) 
plt.bar(x, idle, width=width, label='idle')
for i in range(len(x)):
    x[i] = x[i] + width
plt.bar(x, full, width=width, label='full',tick_label = name_list,fc = 'r')
plt.legend()
plt.xlabel("datacenter")
plt.ylabel("power/kw")
plt.show()
```
