```python
import numpy as np
import pandas as pd

import matplotlib
matplotlib.rcParams['backend'] = 'SVG'
import matplotlib.pyplot as plt

plt.rc('font',family='Times New Roman')
```

```python
df = pd.read_csv("./file/4_11/wind.csv", sep=",")
df = df.drop(['Unnamed: 15'], axis=1)

df_CR = df.loc[df['dcName'] == 'DC_CR'].reset_index()
df_OS = df.loc[df['dcName'] == 'DC_OS'].reset_index()
df_SH = df.loc[df['dcName'] == 'DC_SH'].reset_index()
df_MB = df.loc[df['dcName'] == 'DC_MB'].reset_index()
df_LA = df.loc[df['dcName'] == 'DC_LA'].reset_index()

df_ReSupply_wind = pd.DataFrame({'CR':df_CR['reSupply'], 'MB':df_MB['reSupply'], 'OS':df_OS['reSupply'], 'SH':df_SH['reSupply'], 'LA':df_LA['reSupply']})
```

```python
df = pd.read_csv("./file/4_11/solar.csv", sep=",")
df = df.drop(['Unnamed: 15'], axis=1)

df_CR = df.loc[df['dcName'] == 'DC_CR'].reset_index()
df_OS = df.loc[df['dcName'] == 'DC_OS'].reset_index()
df_SH = df.loc[df['dcName'] == 'DC_SH'].reset_index()
df_MB = df.loc[df['dcName'] == 'DC_MB'].reset_index()
df_LA = df.loc[df['dcName'] == 'DC_LA'].reset_index()

df_ReSupply_solar = pd.DataFrame({'CR':df_CR['reSupply'], 'MB':df_MB['reSupply'], 'OS':df_OS['reSupply'], 'SH':df_SH['reSupply'], 'LA':df_LA['reSupply']})
```

```python
df = pd.read_csv("./file/4_11/total.csv", sep=",")
df = df.drop(['Unnamed: 15'], axis=1)

df_CR = df.loc[df['dcName'] == 'DC_CR'].reset_index()
df_OS = df.loc[df['dcName'] == 'DC_OS'].reset_index()
df_SH = df.loc[df['dcName'] == 'DC_SH'].reset_index()
df_MB = df.loc[df['dcName'] == 'DC_MB'].reset_index()
df_LA = df.loc[df['dcName'] == 'DC_LA'].reset_index()

df_ReSupply_total = pd.DataFrame({'CR':df_CR['reSupply'], 'MB':df_MB['reSupply'], 'OS':df_OS['reSupply'], 'SH':df_SH['reSupply'], 'LA':df_LA['reSupply']})
```

```python
fig = plt.figure(figsize=(10,9),dpi=200)

ax1 = fig.add_subplot(311)
ax1.plot(df_ReSupply_solar['CR'], label='CR', linewidth = '0.8')
ax1.plot(df_ReSupply_solar['MB'], label='MB', linewidth = '0.8')
ax1.plot(df_ReSupply_solar['OS'], label='OS', linewidth = '0.8')
ax1.plot(df_ReSupply_solar['SH'], label='SH', linewidth = '0.8')
ax1.plot(df_ReSupply_solar['LA'], label='LA', linewidth = '0.8')
ax1.set_xlim(0, 1008)
ax1.set_ylim(0, 80000)
ax1.set_title("(a) Solar energy generated within five datacenters for each hour of one week", y = -0.2)
ax1.legend(loc='upper center', bbox_to_anchor=(0.5, 1.2), ncol=5)

ax2 = fig.add_subplot(312, sharex = ax1)
ax2.plot(df_ReSupply_wind['CR'], linewidth = '0.8')
ax2.plot(df_ReSupply_wind['MB'], linewidth = '0.8')
ax2.plot(df_ReSupply_wind['OS'], linewidth = '0.8')
ax2.plot(df_ReSupply_wind['SH'], linewidth = '0.8')
ax2.plot(df_ReSupply_wind['LA'], linewidth = '0.8')
ax2.set_xlim(0, 1008)
ax2.set_ylim(0, 110000)
ax2.set_title("(b) Wind energy generated within five datacenters for each hour of one week", y = -0.2)

ax3 = fig.add_subplot(313, sharex = ax1)
ax3.plot(df_ReSupply_total['CR'], linewidth = '0.8')
ax3.plot(df_ReSupply_total['MB'], linewidth = '0.8')
ax3.plot(df_ReSupply_total['OS'], linewidth = '0.8')
ax3.plot(df_ReSupply_total['SH'], linewidth = '0.8')
ax3.plot(df_ReSupply_total['LA'], linewidth = '0.8')
ax3.set_xlim(0, 1008)
ax3.set_ylim(0, 110000)
ax3.set_title("(c) Renewable energy generated within five datacenters for each hour of one week", y = -0.2)

plt.xticks(np.arange(0, 1152, 144), np.arange(0, 192, 24))
plt.savefig("../pic/4_13/svg/energyGeneration.svg", format='svg')
#plt.show()
```
