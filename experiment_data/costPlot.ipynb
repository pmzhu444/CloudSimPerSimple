```python
names = ['CR', 'OS', 'MB', 'SH', 'LA']
```

```python
df_rr = pd.read_csv("./file/4_11/rr.csv", sep=",")
df_rr = df_rr.drop(['Unnamed: 15'], axis=1)

df_CR_rr = df_rr.loc[df_rr['dcName'] == 'DC_CR'].reset_index()
df_OS_rr = df_rr.loc[df_rr['dcName'] == 'DC_OS'].reset_index()
df_SH_rr = df_rr.loc[df_rr['dcName'] == 'DC_SH'].reset_index()
df_MB_rr = df_rr.loc[df_rr['dcName'] == 'DC_MB'].reset_index()
df_LA_rr = df_rr.loc[df_rr['dcName'] == 'DC_LA'].reset_index()

df_RePrice_rr = pd.DataFrame({'CR':df_CR_rr['rePrice'], 'OS':df_OS_rr['rePrice'], 'MB':df_MB_rr['rePrice'], 'SH':df_SH_rr['rePrice'], 'LA':df_LA_rr['rePrice']})
df_ReDcUsage_rr = pd.DataFrame({'CR':df_CR_rr['reDcUsage'], 'OS':df_OS_rr['reDcUsage'], 'MB':df_MB_rr['reDcUsage'], 'SH':df_SH_rr['reDcUsage'], 'LA':df_LA_rr['reDcUsage']})

for name in names:
    for i in range(len(df_RePrice_rr[name])):
        df_RePrice_rr[name][i] = df_RePrice_rr[name][i] * df_ReDcUsage_rr[name][i]

df_RePrice_rr = df_RePrice_rr / 100

df_rr = df_RePrice_rr.mean(1)
df_rr = pd.DataFrame(df_rr)
df_rr = df_rr.rename(columns = {0:'RR'})
```
